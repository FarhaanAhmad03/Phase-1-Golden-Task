```python
df.index = pd.to_datetime(df.Date)
```

```python
stocks = ['AMZN', 'DPZ', 'BTC', 'NFLX']

for stock in stocks:
    plt.scatter(df.index, df[stock])
    plt.xlabel('Date')
    plt.ylabel(stock)
    plt.title(f'{stock} Closing Price')
    plt.show()
```

```python
plt.figure(figsize=(12,6))
for stock in stocks:
    plt.plot(df.index, df[stock], label=stock)
plt.xlabel('Date')
plt.ylabel('Price')
plt.title('Closing Prices of Stocks')
plt.legend()
plt.show()
```

```python
df['Stock'] = df.apply(lambda row: row.name, axis=1)
plt.boxplot(df.groupby('Stock')[df.columns[1]].apply(list))
plt.xlabel('Stock')
plt.ylabel('Closing Price')
plt.title('Distribution of Closing Price by Stock')
plt.xticks(rotation=45)
plt.show()
```
